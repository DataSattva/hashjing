### 1. Bitcoin Genesis Block — January 3, 2009

The first block of the network (Block 0) was hardcoded into the Bitcoin Core software and forms the root of the entire chain. Its coinbase transaction famously includes the newspaper headline:
*“The Times 03/Jan/2009 Chancellor on brink of second bailout for banks”*
— a manifesto-comment by Satoshi on the collapse of the banking system.

The block minted 50 BTC, but due to a special script, the reward is unspendable. Over 25 million subsequent blocks link to this one via hash-chain ancestry.
([Bitcoin Wiki][1], [Investopedia][2])

### 2. Ethereum Genesis Block — July 30, 2015

Launched at 03:26 UTC, Ethereum’s block 0 established the initial balance of ~60 million ETH, distributed among participants of the 2014 crowdsale and the developer fund. Its hash `0xd4e5…fa3` became the synchronization anchor for all Geth/Parity clients and locked in the network parameters — gas limit, difficulty, and more. From this moment, smart contracts became available on the mainnet, paving the way for DeFi and NFTs. ([Medium][3])

### 3. DAO Fork – Block 1,920,000 (ETH Chain)

On July 17, 2016, the community voted to roll back the theft of 3.6 million ETH from The DAO fund. At block height 1,920,000, a hard fork was implemented: the stolen funds were redirected into a special refund contract. The hash of the main fork block `0x4985…11bb` sealed this “social” decision, and the chain that retained this change became known as **ETH**, the Ethereum we know today. ([Ethereum Stack Exchange][4])

### 9. The “Zero Address” — Where Tokens Are Burned

In Ethereum (and most EVM-compatible blockchains), the address  
`0x0000000000000000000000000000000000000000` exists as a valid recipient — **but it has no private key**. Sending ETH, ERC-20 tokens, or NFTs to this address means that no one — not the sender, miners, or client developers — can ever sign a transaction “back.” That’s why transfers to the zero address are considered **irreversible burns**: the assets are permanently removed from circulation, reducing the circulating supply. ([Cointelegraph][1], [Blockchain Magazine][2])

The earliest “cultural” example is linked to community practices: ERC-20 token supplies were often hard-capped by sending excess tokens to `0x0…0`, since the standard lacked a dedicated `burn()` function for a long time. The same logic applies to NFTs today: if a creator wants to destroy a token, they send it to the zero address (or to the symbolic variant `0x000…dEaD`). In this way, the blockchain’s “black hole” becomes a kind of **crypto-void** — a deterministic graveyard for forgotten projects, marketing airdrops, or intentional sacrifices. ([Ethereum Stack Exchange][3], [Crypto Council for Innovation][4])
